```python
#Average height in men's swimming team
file2=open("Men's swimming average height.txt", "r")
msh=file2.read()
msh1=msh.split()
i=0
avgmsh=0
file2=open("Men's swimming average height in centimeters.txt", "a")
with open("Men's swimming average height in centimeters.txt", "w") as f3:
  while i<len(msh1):
    height1=msh1[i].split("'")
    feet=float(height1[0])
    inches=float(height1[1])
    feet=feet*30.48
    inches=inches*2.54
    msheight2=feet+inches
    i=i+1
    avgmsh+=msheight2/(len(msh1))
    msheight3=str(msheight2)
    f3.write(msheight3)
    f3.write("\n")
    print(msheight2)
print("The average height in centimeters on men's swimming team is: ",avgmsh)
```
